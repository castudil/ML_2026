## Instrucciones de Evaluación de Modelos
Para validar modelos clasificadores es imperativo conocer métricas más allá de la simple exactitud.
Calculen de forma manual (sin librerías como scikit-learn) las siguientes métricas confusionales:
- Exactitud (Accuracy)
- Precisión (Precision)
- Exhaustividad (Recall)
- F-1 Score

```python
import numpy as np

y_true = np.array([1, 0, 1, 1, 0, 1, 0, 0, 1, 0])
probabilidades = np.array([0.8, 0.6, 0.9, 0.3, 0.4, 0.95, 0.2, 0.7, 0.6, 0.1])
y_pred = (probabilidades >= 0.5).astype(int)

def calcular_metricas(y_real, y_prediccion):
    # Implementación Manual
    VP = np.sum((y_real == 1) & (y_prediccion == 1))
    VN = np.sum((y_real == 0) & (y_prediccion == 0))
    FP = np.sum((y_real == 0) & (y_prediccion == 1))
    FN = np.sum((y_real == 1) & (y_prediccion == 0))
    
    acc = (VP + VN) / len(y_real)
    prec = VP / (VP + FP + 1e-10)
    rec = VP / (VP + FN + 1e-10)
    f1 = 2 * (prec * rec) / (prec + rec + 1e-10)
    return acc, prec, rec, f1

acc, prec, rec, f1 = calcular_metricas(y_true, y_pred)
print(f'Accuracy: {acc}, Precision: {prec}, Recall: {rec}, F1: {f1}')
```
