## ¿Qué es la Regresión Logística?
A pesar de contener la palabra "regresión" en su nombre histórico, la **Regresión Logística** es en realidad un algoritmo predictivo para resolver problemas de **clasificación** (típicamente binaria).

Mientras que la regresión lineal intenta predecir un valor numérico a escala continua (como el precio de una casa), la regresión logística predice la **probabilidad** de que un caso pertenezca o no a una categoría determinada (clase $0$ o clase $1$). Dado que estima probabilidades bi-condicionales, el modelo matemáticamente fuerza que su predicción resultante viva estrictamente en el dominio $[0, 1]$. Ejemplos prácticos incluyen: predicción de impago de clientes, clasificar si un pixel es un tejido benigno o maligno, y sistemas anti-spam.

## Limitaciones del MSE y Función Sigmoide
El MSE no es ideal para clasificación pues penaliza predicciones correctas muy extremas y no mapea a probabilidades. Introducimos la Función Sigmoide $\sigma(z)$:

$$ \sigma(z) = \frac{1}{1 + e^{-z}} \tag{1} $$
$$ h_\theta(x) = \sigma(\theta^T x) \tag{2} $$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoide(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 200)
plt.plot(z, sigmoide(z))
plt.title('Función Sigmoide')
plt.grid()
plt.show()
```

## Entropía Cruzada Binaria (Log Loss)
Para clasificación usamos la Pérdida Logarítmica (Log Loss):

$$ J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right] \tag{3} $$

<div class="alert alert-warning">
<b>Profundización Analítica:</b>
<br><br>
Derivación técnica de la función Log Loss a partir del principio de Estimación de Máxima Verosimilitud (MLE).
$$ P(y=1 | x; \theta) = h_\theta(x) \tag{4} $$
$$ p(y | x; \theta) = (h_\theta(x))^y (1 - h_\theta(x))^{1 - y} \tag{5} $$

Log-Verosimilitud:
$$ \ell(\theta) = \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right] \tag{6} $$
Minimizar un costo es maximizar el log-verosimilitud (negativo).

<b>Equivalencia en Regla de Actualización:</b>
Derivando Log Loss respecto a $\theta_j$ (sabiendo que $\sigma'(z) = \sigma(z)(1 - \sigma(z))$):
$$ \frac{\partial}{\partial \theta_j} J(\theta) = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} \tag{7} $$
La regla del gradiente recae idéntica algebraicamente a la del MSE lineal.

</div>

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, random_state=42, class_sep=1.5)
X_b = np.c_[np.ones((len(X), 1)), X]
y = y.reshape(-1, 1)

alpha = 1.0
n_iters = 500
theta = np.zeros((3, 1))

for _ in range(n_iters):
    h = sigmoide(X_b.dot(theta))
    gradiente = (1 / len(y)) * X_b.T.dot(h - y)
    theta = theta - alpha * gradiente

# Plot Frontera
x_dec = np.array([X[:, 0].min(), X[:, 0].max()])
y_dec = -(theta[0] + theta[1]*x_dec) / theta[2]
plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), cmap='bwr')
plt.plot(x_dec, y_dec, 'k--')
plt.title('Regresión Logística - Frontera de Decisión')
plt.show()
```
